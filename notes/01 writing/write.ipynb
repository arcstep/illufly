```python
from textlong.writing import chat, idea, outline, from_outline
from textlong.parser import parse_markdown
from langchain_zhipu import ChatZhipuAI
```

## idea

```python
x = chat(
    ChatZhipuAI(model="glm-4-flash"),
    task="帮我起一个公司名字，我希望基于现在流行的大语言模型AI实现自己研究的算法，为个人或公司提供长文本生成的业务（包括写作、整理资料、策划文案等等）",
    verbose=False
)
```

```python
x
```

## 输出到文件

```python
chat(
    ChatZhipuAI(),
    output_file="小说/起名字.md",
    task="帮我起一个公司名字，我希望基于现在流行的大语言模型AI实现自己研究的算法，为个人或公司提供长文本生成的业务（包括写作、整理资料、策划文案等等）",
    verbose=False
)
```

## outline

```python
ol = outline(
    ChatZhipuAI(),
    task="我希望基于现在流行的大语言模型AI实现自己研究的算法，为个人或公司提供长文本生成的业务（包括写作、整理资料、策划文案等等），帮我策划一下营销推广该怎么做",
    verbose=False
)
```

```python
ol
```

## from_outline

```python
final_md = from_outline(ChatZhipuAI(), completed=ol, verbose=False)
```
