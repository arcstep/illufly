```python
from textlong import idea, outline, from_outline, more_outline
from textlong.parser import parse_markdown
from langchain_zhipu import ChatZhipuAI
```

## idea

```python
%pwd
```

```python
idea(
    ChatZhipuAI(model='glm-4'),
    "我要写100字小说，主题是一个小男孩，故事是穿越修仙逆袭",        
    input = ["小男孩的师傅是一个飞升失败的妖兽", "小男孩的功法叫做‘冰寒决’"],
    knowledge = '小男孩其实是穿越者',
    verbose=True
)
```

```python
idea(
    ChatZhipuAI(),
    "我要写500字小说，主题是一个小孩，故事是穿越修仙逆袭",
    output_file="创意小说.md",
    base_folder="小说",
    verbose=True
)
```

## outline

```python
ol = outline(
    ChatZhipuAI(model='glm-4'),
    "我要写1000字小说，帮我写一份提纲，主题是一个小孩，故事是穿越修仙逆袭",
    output_file="提纲.md",
    base_folder="小说",
    verbose=True
)
```

```python
ol.to_dict()
```

## from_outline

```python
final_md = from_outline(
    ChatZhipuAI(),
    input="小提纲.md",
    output_file="正文.md",
    base_folder="小说",
    verbose=True
)
```

```python
print(final_md.output_text)
```

## outline_from_outline

```python
new_ol = more_outline(
    ChatZhipuAI(),
    input="小提纲.md",
    output_file="更大的提纲.md",
    base_folder="小说",    
    verbose=True
)
```

```python
print(new_ol.output_text)
```
