### 异步调用

```python
# 异步输出
from illufly.chat import ChatOpenAI
from illufly.io import alog

a = ChatOpenAI(model="gpt-3.5-turbo")
await a(
    "你能帮我写一首关于兔子做梦的四句儿歌?",
    verbose=True,
    handlers=[alog]
)
a.memory
```

```python
# 在调用时指定模型名称
await ChatOpenAI(handlers=[alog])(
    "你能帮我写一首关于兔子做梦的四句儿歌?",
    model="gpt-3.5-turbo",
    temperature=0.1
)
a.memory
```

### 工具回调

```python
from illufly.chat import ChatOpenAI
from illufly.types import BaseAgent

def get_current_weather(location: str):
    """获取城市的天气情况"""
    yield f"{location}今天是晴天。 "
    
a = ChatOpenAI(tools=[BaseAgent(get_current_weather)])
a("今天广州出门要准备什么", new_chat=True, verbose=True)
a.memory
```

```python
from illufly.chat import ChatOpenAI
from illufly.types import BaseAgent

def get_current_weather(location: str):
    """获取城市的天气情况"""
    yield f"{location}今天是晴天。 "
    
a = ChatOpenAI(imitator="QWEN", model="qwen-plus", tools=[BaseAgent(get_current_weather)])
a("今天广州出门要准备什么", new_chat=True)
a.memory
```

```python
from illufly.chat import ChatOpenAI
from illufly.io import log, usage

def get_current_weather(location: str):
    """获取城市的天气情况"""
    return f"{location}今天是晴天， 25摄氏度。 "

def get_prices(hours: int):
    """预测订羽毛球球场的价格，hours是预订球场的小时数，因为球场价格是按照小时计算的"""
    return f"球场每小时25元，你要定{hours}小时，所以是{hours*25}元。"
```

```python
a = ChatOpenAI(tools=[get_current_weather, get_prices])
a("帮我看看广州天气，顺便看看订2小时的羽毛球球场多少钱")
a.memory
```

```python
from illufly.types import PromptTemplate

a = ChatOpenAI(memory=PromptTemplate("TOOLS_CALLING"), tools=[get_current_weather, get_prices])
a("帮我看看广州天气，顺便看看订2小时的羽毛球球场多少钱")
a.memory
```
