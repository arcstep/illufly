```python
from illufly.embeddings import HuggingFaceEmbeddings

e = HuggingFaceEmbeddings()
resp = e.embed_documents(["illufly 是一个专注于写作和对话场景的多智能体框架"])
print(resp)
```

```python
from illufly.embeddings import HuggingFaceEmbeddings

e = HuggingFaceEmbeddings()
resp = e.query("illufly 是一个专注于写作和对话场景的多智能体框架")
print(resp)
```
