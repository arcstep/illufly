### 异步调用

```python
from illufly.io import alog
await alog(zhipu, "能改为6句？")
zhipu.memory
```

### 工具回调

```python
from illufly.types import BaseAgent
from illufly.chat import ChatZhipu
from illufly.io import alog

def get_current_weather(location: str, **kwargs):
    """获取城市的天气情况"""
    return f"{location}今天是晴天。 "

zhipu = ChatZhipu(
    tools=[BaseAgent(get_current_weather)],
    model="glm-4"
)
await zhipu("请使用工具查询，今天广州天气如何啊", handlers=[alog], verbose=True)
```

```python
zhipu.memory
```

## 向量模型

```python
from illufly.rag import ZhipuEmbeddings, TextEmbeddings

e = ZhipuEmbeddings()
resp = e.embed_documents(["illufly 是一个专注于写作和对话场景的多智能体框架"])
print(resp)
```

```python
from illufly.embeddings import ZhipuEmbeddings, TextEmbeddings
from illufly.io import log

e = ZhipuEmbeddings()
resp = e.query("illufly 是一个专注于写作和对话场景的多智能体框架")
print(resp)
```
