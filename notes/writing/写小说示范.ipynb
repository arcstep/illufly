## outline

```python
ol = outline(
    ChatZhipuAI(model='glm-4'),
    "我要写1000字小说，帮我写一份提纲，主题是一个小孩，故事是穿越修仙逆袭",
    is_fake=True
)
```

```python
ol = outline(
    ChatZhipuAI(model='glm-4'),
    "我要写1000字小说，帮我写一份提纲，主题是一个小孩，故事是穿越修仙逆袭",
    verbose=True
)
```

```python
print(ol)
```

## from_outline

```python
final_md = from_outline(ChatZhipuAI(), ol, verbose=True, config={"prev_k":10})
```

```python
print(final_md)
```
