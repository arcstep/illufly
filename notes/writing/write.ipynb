```python
def extract_text(resp_md: str, start_marker: str=None, end_marker: str=None):
    """
    提取start_marker和end_markder中间的文字。
    """
    _start_marker = start_marker or '<OUTLINE>'
    _end_marker = end_marker or '</OUTLINE>'
    
    start_index = resp_md.find(_start_marker)
    end_index = resp_md.rfind(_end_marker)
    
    if start_index != -1 and end_index != -1 and start_index < end_index:
        start_index += len(_start_marker)
        return resp_md[start_index:end_index].strip()
    else:
        return resp_md

extract_text("""
<OUTLINE>
为了提升"文思泉涌"的品牌知名度和市场占有率，我们制定了线上线下相结合的营销策略。线上推广方面，我们将在社交媒体平台如微博、知乎、微信公众号上发布结合热点事件或话题的文案，以吸引用户关注。同时，通过合作伙伴的网站和社交媒体渠道进行联合推广，扩大品牌影响力。

线下活动方面，我们会定期举办研讨会和工作坊，邀请行业专家和用户共同参与，通过实际操作演示"文思泉涌"的功能和效果，增加用户对产品的认知度和信任度。此外，参与行业展会和论坛也是我们的重要策略，可以直接与潜在客户接触，收集反馈，优化产品。

对于推广渠道的分析，社交媒体推广的优势在于低成本和快速传播，但缺点是信息量大，用户注意力分散，因此我们建议通过精准定位用户兴趣点和制作高质量内容来提高吸引力。线下活动的优点是互动性强，用户印象深，但成本较高，因此需要精心策划，确保投入产出比。

优化建议包括：在线上，采用数据分析工具，精准投放广告；线下，通过预登记和会后跟踪，提高转化率。结合热点事件营销，我们可以迅速提升品牌曝光度，如在重大科研成果发布时推出相关主题的推广活动，以提升产品的权威性和时效性。
</OUTLINE>
## 客户案例与用户评价""")
```
