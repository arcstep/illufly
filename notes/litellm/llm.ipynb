```python
from illufly.llm import LiteLLM

llm = LiteLLM(model="gpt-4o-mini")
resp = llm.completion(messages="你是什么模型？")
if resp:
    print(resp.choices[0].message.content)
```

```python
from illufly.llm import LiteLLM

llm = LiteLLM(imitator="OPENAI", model="gpt-4o-mini")
resp = llm.completion(messages="你是什么模型？")
if resp:
    print(resp.choices[0].message.content)
```

```python
from illufly.llm import LiteLLM

llm = LiteLLM(imitator="ZHIPU", model="glm-4-flash")
resp = await llm.acompletion(messages="你是什么模型？")
if resp:
    print(resp.choices[0].message.content)
```

```python
from illufly.llm import LiteLLM

llm = LiteLLM(imitator="ZHIPU", model="glm-4-flash")
resp = llm.completion(messages="你是什么模型？", stream=True)
for x in resp:
    if x.choices and x.choices[0].delta.content:
        print(x.choices[0].delta.content, end="")
```

```python
from illufly.llm import LiteLLM, litellm

llm = LiteLLM(imitator="ZHIPU", model="glm-4-flash")
resp = await llm.acompletion(messages="你是什么模型？", stream=True)
async for x in resp:
    if x.choices and x.choices[0].delta.content:
        print(x.choices[0].delta.content, end="")
```

```python
from illufly.llm import LiteLLM, litellm

litellm.enable_cache(
    type="disk",
    disk_cache_dir="__LITE_LLM_CACHE",
    supported_call_types=["completion", "acompletion", "embedding", "aembedding", "atranscription", "transcription"]
)

llm = LiteLLM(imitator="OPENAI")
resp = llm.embedding(input=["你是什么模型？"], model="text-embedding-3-small")
print(resp)
```
