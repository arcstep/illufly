```python
from textlong.md import toolkits, chains

def call_chain(chain, input):
    text = ""
    for chunk in chain.stream(input):
        text += chunk.content
        print(chunk.content, end="")
    
    print(f"\n\n实际字数: {len(text)}")
    return text
```

## Tools-Calling

```python
from textlong.tools import toolkits, chains
from langchain_zhipu import ChatZhipuAI
from langchain_core.utils.function_calling import convert_to_openai_tool
from textlong.tools import toolkits, chains

[convert_to_openai_tool(t) for t in toolkits[:2]]
```

```python
llm = ChatZhipuAI()

def call_llm(input):
    text = ""
    for chunk in llm.stream(input, tools=[convert_to_openai_tool(t) for t in toolkits[:2]]):
        print(chunk)
        text += chunk.content
        # print(chunk.content, end="")
    
    print(f"\n\n实际字数: {len(text)}")
    return text
```

```python
call_llm("创作50字的python教程")
```
